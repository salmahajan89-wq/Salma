## IMPORTAZIONE FILE CSV SU MYSQL PER POPOLARE TABELLA
Per le tabelle di piccola dimensione, i file sono stati importati direttamente da 'IMPORT' di Mysql, mentre Reviews viene importato con la pipeline

```python
# Importazione librerie necessarie
import pandas as pd
from sqlalchemy import create_engine
```

```python
#  PASSAGGIO 1: Configurazione Connessione
DB_NAME = 'Hotel_Reviews' 
engine = create_engine(f'mysql+pymysql://root:@127.0.0.1:3306/{DB_NAME}')

# PASSAGGIO 2: Pipeline a blocchi 
file_path = 'Reviews.csv'

def avvia_caricamento():
    try:
        print(f"Inizio caricamento di {511944} righe...")
        
        # Caricamento a blocchi
        for i, chunk in enumerate(pd.read_csv(file_path, chunksize=50000, low_memory=False)):
            # Caricamento del blocco nella tabella 'Reviews'
            chunk.to_sql('Reviews', con=engine, if_exists='append', index=False)
            
            # Esecuzione calcolo per mostrare il progresso
            righe_caricate = (i + 1) * 50000
            if righe_caricate > 511944: righe_caricate = 511944
            print(f"Stato: {righe_caricate} / 511944 righe caricate...")

        print("\nCARICAMENTO COMPLETATO CON SUCCESSO!")

    except Exception as e:
        print(f"\nERRORE DURANTE IL CARICAMENTO: {e}")

if __name__ == "__main__":
    avvia_caricamento()

```
