```python
H_R_C = Hotel_Rev.copy()
```

# DATA CLEANING

### GESTIONE MISSING VALUES

```python
#Cancellazione delle righe con i valori nulli 
H_R_C.dropna(inplace=True)

#controllo della pulizia se avvenuta correttamente
H_R_C.isnull().sum()

```

### CONVERSIONE IN DATETIME ED ESTRAZIONE ANNO/MESE

```python
# Conversione colonna 'Review_Date' da str a Datetime
H_R_C['Review_Date'] = pd.to_datetime(H_R_C['Review_Date'])

#Estrazione dell'anno
H_R_C['Year'] = H_R_C['Review_Date'].dt.year

#Estrazione del mese
H_R_C['Month'] = H_R_C['Review_Date'].dt.month

#Verificazione dell'avvenuta modifica
print(H_R_C['Review_Date'].dtype)
print(H_R_C[['Review_Date', 'Year', 'Month']].head(4))

```

```python
#Rinominazione colonne 

H_R_C = H_R_C.rename(columns={
    'Total_Number_of_Reviews_Reviewer_Has_Given': 'Reviewer_Total_Reviews'})
```

### RIMOZIONE DUPLICATI ESATTI

```python
# cancellazione delle righe con valori dupplicati
H_R_C.drop_duplicates(inplace=True)

# controllo della corretta avvenuta cancellazione
H_R_C.duplicated().sum()

```

### CREAZIONE VARIABILI DERIVATE

```python
# Creazione prima colonna derivata dalla colonna 'Tags' tramite una funzione che in base
# alla parola chiave rilevata nella tag associa la recensione a una determinata categoria:

def identificazione_ospite(tag):
    tag_lower = str(tag).lower()
    
    if 'family' in tag_lower:
        return 'Family'
    elif 'couple' in tag_lower:
        return 'Couple'
    elif 'group' in tag_lower:
        return 'Group'
    elif 'solo' in tag_lower:
        return 'Solo'
    else:
        return 'Other'
    
# Creazione della nuova colonna
H_R_C['Guest_Category'] = H_R_C['Tags'].apply(identificazione_ospite)
```

```python
# Creazione seconda colonna derivata dalla colonna 'Hotel_Address' tramite una funzione
# che in base alla ultima/penultima parola rilevata nell'indirizzo, restituisce nella nuova
# colonna il paese di appartenenza dell'hotel.

def estrazione_paese(indirizzo):
    parola = indirizzo.split()
    if "Kingdom" in indirizzo:
        return " ".join(parola[-2:])
    else:
        return parola[-1]
    
H_R_C['Hotel_Country'] = H_R_C['Hotel_Address'].apply(estrazione_paese)
    
```

```python
# Creazione  terza colonna derivata dalla colonna 'Hotel_Address' tramite una funzione
# che in base alla parola precedente al pease, rilevata nell'indirizzo, restituisce nella nuova
# colonna la città di appartenenza dell'hotel

def estrazione_città(row):
    indirizzo = row['Hotel_Address']
    paese = row['Hotel_Country']
    
    # 1. Rimozione del nome del paese dall'indirizzo e degli spazi extra
    indirizzo_pulito = indirizzo.replace(paese, '').strip()
    parole = indirizzo_pulito.split()
    
    if paese == 'United Kingdom':
        # selezione delle ultime 3 parole (es: 'London W2 3NA') 
        # e tra queste tre la prima (London)
        if len(parole) >= 3:
            return parole[-3]
        return parole[0] # Fallback se l'indirizzo è corto
    else:
        # Per le altre nazioni: selezione dell'ultima parola rimasta dopo la rimozione del paese
        return parole[-1]

# Applicazione della funzione usando 'axis=1' poichè occorre sia Indirizzo che Paese
H_R_C['Hotel_City'] = H_R_C.apply(estrazione_città, axis=1)

```

```python
# Esecuzione controllo per verificare che abbia associato alla nuova colonna
# solo valori corrispondenti a paesi e città nelle rispettive colonne:
print(H_R_C['Hotel_Country'].unique())
print(H_R_C['Hotel_City'].unique())
```

```python
#Salvataggio di una copia in formato CSV del df manipolato
H_R_C.to_csv('hotel_reviews_clean.csv', index=False)
```

```python
#Creazione db Hotels
Hotels = H_R_C[['Hotel_Name', 'Hotel_Address', 'Hotel_Country', 'Hotel_City']].drop_duplicates().reset_index(drop=True)
Hotels.insert(0, 'Hotel_ID', ['H' + str(i) for i in range(1, len(Hotels) + 1)])

Hotels.to_csv('Hotels.csv', index=False)
```

```python
#Creazione db Reviewers
Reviewers = H_R_C[['Reviewer_Nationality', 'Guest_Category', 'Reviewer_Total_Reviews']].drop_duplicates().reset_index(drop=True)
Reviewers.insert(0, 'Reviewer_ID', ['Rr' + str(i) for i in range(1, len(Reviewers) + 1)])

Reviewers.to_csv('Reviewers.csv', index=False)
```

```python
#Creazione db Reviews

Reviews = H_R_C.copy()

colonne_identità_reviewer = ['Reviewer_Nationality', 'Guest_Category', 'Reviewer_Total_Reviews']
colonne_identità_hotel = ['Hotel_Name', 'Hotel_Address', 'Hotel_Country', 'Hotel_City']

# Collegamento dell'Hotel_ID corretto tramite il nome dell'hotel
Reviews = Reviews.merge(Hotels, on=colonne_identità_hotel, how='left')

# Collegamento del Reviewer_ID tramite la nazionalità (o altra colonna identificativa)
Reviews = Reviews.merge(Reviewers, on=colonne_identità_reviewer, how='left')

# Creazione l'ID univoco per ogni singola recensione
Reviews.insert(0, 'Review_ID', ['Rw' + str(i) for i in range(1, len(Reviews) + 1)])

```

```python
#Creazione db Hotel_Stats
Hotel_Stats = Reviews.copy()
Hotel_Stats = Hotel_Stats.drop(columns=['Positive_Review',	'Reviewer_Total_Reviews',	'Reviewer_Score', 'Review_ID', 'Hotel_Address', 'Review_Date',	'Hotel_Name',	'Reviewer_Nationality',	'Negative_Review',	'Review_Total_Negative_Word_Counts',	'Tags',	'Year',	'Month',	'Guest_Category',	'Hotel_Country',	'Reviewer_ID', 'Hotel_City'])

```

```python
H_S = Hotel_Stats.copy()




# 2. Creazione di una colonna temporanea numerica estraendo solo i numeri da '100 day'
H_S['days_numeric'] = H_S['days_since_review'].str.extract('(\d+)').astype(int)

# 3. Ordine del DataFrame: prima per Hotel_ID e poi per i giorni in modo decrescente
H_S_sorted = H_S.sort_values(by=['Hotel_ID', 'days_numeric'], ascending=[True, False])

# 4. Rimozione dei duplicati mantenendo solo la prima occorrenza (la più recente)
H_S_pulito = H_S_sorted.drop_duplicates(subset=['Hotel_ID'], keep='first')

# 5. Rimuozione dela colonna di supporto numerica per tornare alla struttura originale
H_S_pulito = H_S_pulito.drop(columns=['days_numeric'])

H_S_pulito.sort_values(by=['Hotel_ID'])

H_S_pulito.to_csv('Hotel_Stats.csv', index=False)

print(H_S_pulito.head(10))
```
